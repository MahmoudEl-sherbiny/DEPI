## Mushroom Classification Project

### Problem Description
The goal of this project is to classify mushrooms as **edible (e)** or **poisonous (p)** based on 20 descriptive features such as cap shape, cap color, stem characteristics, habitat, and season.  
This is a **binary classification problem** where the target variable is the mushroom class (`edible` or `poisonous`).

```python
from sklearn.feature_selection import f_classif, mutual_info_classif

features_data = data[["cap-diameter", "stem-height", "stem-width"]]

target = data['class'].str.lower().map({'e': 0, 'p': 1}).astype(int)
f_scores, p_vals = f_classif(features_data, target)
mi_scores = mutual_info_classif(features_data, target, discrete_features=False, random_state=0)

num_results = pd.DataFrame({
    'Feature': features_data.columns,
    'F_score': f_scores,
    'p_value': p_vals,
    'Mutual_Info': mi_scores
}).sort_values(['F_score','Mutual_Info'], ascending=False)

print(num_results)
```

```python
from scipy.stats import pointbiserialr

pb_rows = []
for col in features_data.columns:
    r, p = pointbiserialr(target, data[col])
    pb_rows.append({'Feature': col, 'PointBiserial_r': r, 'p_value': p})

pb_results = pd.DataFrame(pb_rows).sort_values('PointBiserial_r', ascending=False)
print(pb_results)
```

```python
from scipy.stats import chi2_contingency

cat_cols = data.select_dtypes(include=['object']).columns.drop('class')

y = data['class']

def cramers_v(conf_mat: pd.DataFrame) -> float:
    chi2, _, _, _ = chi2_contingency(conf_mat)
    n = conf_mat.values.sum()
    r, k = conf_mat.shape
    return np.sqrt((chi2 / n) / (min(k - 1, r - 1)))


cat_results = []

for col in cat_cols:
    s = data[col].fillna('Missing')

    top = s.value_counts().head(30).index
    s = s.where(s.isin(top), 'Other')

    ct = pd.crosstab(s, y)

    chi2, p, dof, _ = chi2_contingency(ct)
    cv = cramers_v(ct)

    cat_results.append({
        'Feature': col,
        'Levels': ct.shape[0],
        'Chi2_pvalue': p,
        'CramersV': cv
    })

cat_results = pd.DataFrame(cat_results).sort_values('CramersV', ascending=False)
print(cat_results)
```

# handle missing data in cateogrical columns on train data

```python
for col in ["cap-surface", "gill-attachment"]:
    mode_value = X_train[col].mode()[0]
    X_train[col].fillna(mode_value, inplace=True)
    X_test[col].fillna(mode_value, inplace=True)
```

```python
# Make scaling for Numeric features and by Robust Scalling
from sklearn.preprocessing import RobustScaler


scaler_obj = RobustScaler()
X_train_scaled = scaler_obj.fit_transform(X_train[numerical_col])
X_test_scaled = scaler_obj.transform(X_test[numerical_col])

X_train_scaled = pd.DataFrame(X_train_scaled, columns=numerical_col, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=numerical_col, index=X_test.index)

# Replace original numeric columns with scaled ones
X_train = X_train.drop(columns=numerical_col).join(X_train_scaled)
X_test = X_test.drop(columns=numerical_col).join(X_test_scaled)
```

```python
# from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['cap-shape', 'cap-surface', 'cap-color',
                    'does-bruise-or-bleed', 'gill-attachment', 'gill-color',
                    'stem-color', 'has-ring', 'ring-type', 'habitat', 'season']

X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True).reindex(columns=X_train.columns, fill_value=0)
```

### LogisticRegression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

logistic_model = LogisticRegression(class_weight="balanced", max_iter=1000)
logistic_model.fit(X_train, y_train)
# print(logistic_model.score(X_test, y_test)) it's accuracy he make y_pred and calc Accuracy under the hood
y_pred = logistic_model.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {'C': [0.01, 0.1, 1, 10, 100]}
grid = RandomizedSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best C:", grid.best_params_)
print("Best CV score:", grid.best_score_)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
# from sklearn.metrics import confusion_matrix, classification_report

from sklearn.model_selection import RandomizedSearchCV
params = {"n_neighbors": [1, 3, 5, 7, 9, 11]}
grid = RandomizedSearchCV(KNeighborsClassifier(), params, cv=5, scoring="accuracy")

grid.fit(X_train, y_train)

print("Best k:", grid.best_params_)
print("Best CV score:", grid.best_score_)

best_knn = grid.best_estimator_
test_acc = best_knn.score(X_test, y_test)
print("Test Accuracy:", test_acc)
```

```python
from sklearn.metrics import accuracy_score

k_values = [1, 3, 5, 7, 9, 11]
accuracies = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

plt.plot(k_values, accuracies, marker='o')
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Test Accuracy")
plt.title("KNN Performance with Different k")
plt.grid(True)
plt.show()
```
