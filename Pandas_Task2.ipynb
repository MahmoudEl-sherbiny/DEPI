# üêº Pandas Practice: DataFrames, Selection, Mapping, and More


```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Sample data
names = ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan', 'Fiona', 'George', 'Hannah']
departments = ['Sales', 'Marketing', 'HR', 'Tech']

# Create a DataFrame
df = pd.DataFrame({
    'Name': np.random.choice(names, size=20),
    'Age': np.random.randint(22, 60, size=20),
    'Department': np.random.choice(departments, size=20),
    'Salary': np.random.randint(40000, 120000, size=20),
    'YearsExperience': np.random.randint(0, 20, size=20)
})
df.head()
```

# 1Ô∏è‚É£ Use .loc to select all rows where the Department is 'Tech'.
#    - Describe what .loc does and why it's useful for label-based filtering.


```python
df.loc[df["Department"]== 'Tech']
```

.loc : It allows you to select rows and columns by their labels (names)

Why it useful?
1. Readable: It makes your code more intuitive by using **column names** and row labels directly.

2. Flexible: You can **filter rows and select specific columns at the same time**
3. Supports complex conditions


# 2Ô∏è‚É£ Use .iloc to select the first 5 rows and the last two columns.
#    - What is the difference between .loc and .iloc?


```python
df.iloc[:5, -2:]
```

# .loc   (Label-Based Filtering)
1. Based on Labels "Names"
2. Can use Condition
3. df.loc[row_labels, col_labels]

# .iloc  (integer-position-based access)
1. Based on Integer  Positions "Index numbers"
2. Not directly with boolean conditions
3. df.iloc[row_positions, col_positions]



# 3Ô∏è‚É£ Map a new column called 'DeptCode' where:
#     'Sales' -> 1, 'Marketing' -> 2, 'HR' -> 3, 'Tech' -> 4.
#    - Use .map and explain what happens if a value is not mapped.


```python
dept_mapping = {
    'Sales': 1,
    'Marketing': 2,
    'HR': 3,
    'Tech': 4
}

df['DeptCode'] = df['Department'].map(dept_mapping)
```

.map() will assign NaN to that row in 'DeptCode' if a value isnot mapped.


# 4Ô∏è‚É£ Use .apply to calculate a new column 'Seniority' where:
#     - If YearsExperience > 10 ‚Üí 'Senior'
#     - If between 5-10 ‚Üí 'Mid-Level'
#     - Else ‚Üí 'Junior'
#    - Use a lambda function with apply.


```python
df["Seniority"] = df["YearsExperience"].apply(lambda x: "Senior" if x > 10 else "Mid-Level" if x in range(5,11) else "Junior" )
df
```


# 1Ô∏è‚É£4Ô∏è‚É£ Use .apply to normalize the Salary column (min-max scaling between 0 and 1).
#     - Bonus: Write your own normalization function and pass it to apply.


```python
def min_max_nomralization(value):
  return (value - df["Salary"].min())  / (df["Salary"].max() - df["Salary"].min())

df["Normalize_Salary"] = df["Salary"].apply(min_max_nomralization)
df
```
